## transformations

### Hisogram

```python

###### Hisogram

# the following functions return images after transformation, and save their corresponding histogram

# Function for color image histogram
def color_image_histogram(image):
    #display image in notebook 
    plt.imshow(image) 
    plt.show()
    #Get min and max grayscale values
    mini=np.min(image)
    maxi=np.max(image)
    bins = maxi - mini
    # Create figure to save the histogram
    fig = plt.figure()
    # calculate histogram for each color channel
    for i, col in enumerate(['r', 'g', 'b']):
        # cv2.calcHist(images, channels, mask, bins, ranges)
        hist = cv2.calcHist([image], [i], None, [256], [0, 256])        
        plt.plot(hist, color = col)
        plt.xlim([mini, maxi])  
    # save calculated histogram image
    fig.savefig('hist.png')
    plt.show()
    
    return image
    

image = io.imread('SampleImage1.jpg')
color_image_histogram(image)
    
    
# Function for gray-scale histogram
def gray_image_histogram(gray):
    #display image 
    plt.imshow(gray, cmap='gray') 
    plt.show()
    # Create figure to save the histogram
    fig = plt.figure()
    #Get min and max grayscale values
    mini=np.min(gray)
    maxi=np.max(gray)
    bins = maxi - mini
    # calculate histogram for grayscale channel
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256])        
    plt.plot(hist, color = 'k')
    plt.xlim([mini, maxi])
    # save calculated histogram image
    fig.savefig('hist.png')
    plt.show()
    return gray
    
# gray = cv2.imread('SampleImage2.gif')
gray = io.imread('SampleImage2.gif')
gray_image_histogram(gray)
```

## transformation 

### 1. linear stretching

```python
# 1. linear stretching

# Function for grayscale image linear Stretching
def grayscale_linearStretching(gray):
    # rescale_intensity(image, in_range='image', out_range='dtype')
    #in_range and out_range, are used to stretch or shrink the intensity range of the input image.
    # inrange=(0, 255), outrange=(0, 255)
    # better to set them to image default 
    stretched_grayimage = rescale_intensity(gray,in_range='image', out_range='image')
    # get histogram by calling gray_image_histogram() function
    gray_image_histogram(stretched_grayimage)
    return stretched_grayimage



grayscale_linearStretching(gray)    


# Function for color image linear Stretching
def color_linearStretching(image):
    channels = cv2.split(image)
    st_channels = []
    # Apply linear Stretching for each color channel
    for ch, color in zip(channels, ['R', 'G', 'B']):
        st_channels.append(rescale_intensity(ch))
    stretched_image = cv2.merge(st_channels)
    color_image_histogram(stretched_image)
    return stretched_image


color_linearStretching(image)

```

### 2. histogram equalization 


```python

# 2. histogram equalization 

# Function for grayscale image equalization
def grayscale_equalization(gray):
    equalized_grayimage = cv2.equalizeHist(gray)
    gray_image_histogram(equalized_grayimage)
    return equalized_grayimage
    
grayscale_equalization(gray)    


# Function for color image equalization
def color_equalization(image):
    # split image channels and equalize each seperatly then combine them.
    channels = cv2.split(image)
    eq_channels = []
    for ch, color in zip(channels, ['R', 'G', 'B']):
        eq_channels.append(cv2.equalizeHist(ch))
    eq_image = cv2.merge(eq_channels)
    
# Other possible method by splitting the channel, equalizing V channel, then combining them togrther
# By experminting, I found that the first method produce better results.
#     H, S, V = cv2.split(cv2.cvtColor(image, cv2.COLOR_RGB2HSV))
#     eq_V = cv2.equalizeHist(V)
#     eq_image = cv2.cvtColor(cv2.merge([H, S, eq_V]), cv2.COLOR_HSV2RGB)

    color_image_histogram(eq_image)
    
    return eq_image


color_equalization(image)
```

### 3. power transformation

```python
# 3. power transformation

# Function for grayscale image power trans
def grayscale_power(gray, gamma,gain):
    power_grayimage = adjust_gamma(gray,gamma,gain)
    gray_image_histogram(power_grayimage)
    return power_grayimage

gain =1    
gamma =2
grayscale_power(gray,gamma,gain)    


# Function for color image power trans
def color_power(image, gamma,gain):
    channels = cv2.split(image)
    pw_channels = []
    for ch, color in zip(channels, ['R', 'G', 'B']):
        pw_channels.append(adjust_gamma(ch,gamma,gain))
    power_image = cv2.merge(pw_channels)
    color_image_histogram(power_image)
    return power_image


color_power(image,gamma,gain)


```

```python
# Function to select image, and apply the selected transformation to it

def select_image():
    # image panels
    global panelA, panelB,panelC,panelD
    
    # Allow user to select an image and store its path
    path = filedialog.askopenfilename()

    # check if a file is selected
    if len(path) > 0:

        # check image color or grey scale
        scale = is_grey_scale(path)
        
        # load the image 
        image = io.imread(path)  
        # Get image histograms and process it whith choosen process 
        clicked("Histogram",image,scale) 
        original_hist = io.imread("hist.png")
        processd = clicked(choice.get(),image,scale) 
        processed_hist = io.imread("hist.png")
        
        # convert the images to PIL format
        image = Image.fromarray(image)
        processd = Image.fromarray(processd)
        original_hist =Image.fromarray(original_hist)
        processed_hist =Image.fromarray(processed_hist)

        # then to ImageTk format
        image = ImageTk.PhotoImage(image)
        processd = ImageTk.PhotoImage(processd)
        original_hist =ImageTk.PhotoImage(original_hist)
        processed_hist =ImageTk.PhotoImage(processed_hist)
        
        
        
        # if the panels are None, initialize them
        if panelA is None or panelB is None or panelC is None or panelB is None:
            # the first panel will store our original image
            panelA = Label(leftframe,image=image)
            panelA.image = image
            panelA.pack( padx=10, pady=10)
            
            #
            panelC = Label(leftframe,image=original_hist)
            panelC.image = original_hist
            panelC.pack( padx=10, pady=10)
            
            # while the second panel will store the processd

            panelB = Label(rightframe, image=processd)
            panelB.image = processd
            panelB.pack(padx=10, pady=10)
            
            #
            panelD = Label(rightframe, image=processed_hist)
            panelD.image = processed_hist
            panelD.pack( padx=10, pady=10)
        # else, update image panels
        else:
            # update the pannels
            panelA.configure(image=image)
            panelB.configure(image=processd)
            panelC.configure(image=original_hist)
            panelD.configure(image=processed_hist)
            panelC.image = original_hist
            panelD.image = processed_hist
            panelA.image = image
            panelB.image = processd

# initialize window toolkit and image panels
root = Tk()
panelA = None
panelB = None
panelC = None
panelD = None

root.wm_title("Image Processing")

# initialize frame
frame = Frame(root)
frame.pack()
 
#Create left frame
leftframe = Frame(frame)
leftframe.pack(side=LEFT)

#Create right frame
rightframe = Frame(frame)
rightframe.pack(side=RIGHT)

#Create top frame
topframe = Frame(frame)
topframe.pack(side=TOP, pady=5, padx=5)

#Add a label to top 
mainLabel = Label(topframe, text = "Image Processing",height=1, width=50 )
mainLabel.pack(side="top")


###### Create Tasks List

thetext = "Choose a transformation \n to apply on the image:"    
tasksLabel = Label(topframe, text = thetext,height=3, width=60 )
tasksLabel.pack(padx = 5, pady = 15)   

transformations = [
    ("Histogram", "Histogram"),
    ("Histogram Equalization", "Histogram Equalization"),
    ("Linear Streatching", "Linear Streatching"),
    ("Power Transformation", "Power Transformation")]

choice = StringVar()
choice.set("Histogram")

for text, task in transformations:
    Radiobutton(topframe, text=text, variable=choice, value=task).pack()


    
# create a button, then when pressed, will trigger a file chooser
# dialog and allow the user to select an input image; then add the
# button the GUI
btn = Button(topframe, text="Select an image to process", command=select_image)
btn.pack(side="bottom", padx="10", pady="50")



# Get parameters input from user  
inputLabel = Label(topframe, text="with gamma value: ")
e1 = Entry(topframe)
inputLabel.pack(side="top", pady=0, padx=5)
# e1.insert(0, 0.2) 
e1.pack(side="top", pady=0, padx=5)




# Function to check selected processing & apply it

R_Label = None
L_Label = None
def clicked(value,image,scale):
    global myLabel,R_Label

    if R_Label is None:
        R_Label = Label(rightframe, text="Image After " + value)
        R_Label.pack(padx = 5, pady = 5)
        L_Label = Label(leftframe, text="Original Image ")
        L_Label.pack(padx = 5, pady = 5)
    else:
        R_Label.configure(text="Image After " + value)
        R_Label.text = text="You selected " + value
    
    if scale == True:
        if value == "Histogram":
            processd = gray_image_histogram(image)
        elif value == "Histogram Equalization":
            processd = grayscale_equalization(image)     
        elif value == "Linear Streatching":
            processd = grayscale_linearStretching(image)    
        elif value == "Power Transformation":
            if e1.get() != '':
                gammaValue = float(e1.get())
            processd = grayscale_power(image,gammaValue,1)    
        return processd
    else:
        if value == "Histogram":
            processd = color_image_histogram(image)
        elif value == "Histogram Equalization":
            processd = color_equalization(image)     
        elif value == "Linear Streatching":
            processd = color_linearStretching(image)    
        elif value == "Power Transformation":
            if e1.get() != '':
                gammaValue = float(e1.get())
            processd = color_power(image,gammaValue,1)    
        return processd       
        

# Function to check if image is grey scale

def is_grey_scale(img_path):
    img = Image.open(img_path).convert('RGB')
    w, h = img.size
    for i in range(w):
        for j in range(h):
            r, g, b = img.getpixel((i,j))
            if r != g != b: 
                return False
    return True


# kick off the GUI
root.mainloop()
```
